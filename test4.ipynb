```python

# Q.1. Write a Python program to remove all duplicate elements from a given array and returns a new array. 
# Sample Output:
# Original array: 1 3 5 1 3 7 9
# After removing duplicate elements : 1 3 5 7 9
# Original array: 2 4 2 6 4 8
# After removing duplicate elements : 2 4 6 8

def remove_dup(n):
    a=set()
    for i in n:
        a.add(i)
    return list(a)    
ar=[1,3,5,1,3,7,9]
ar1=[2,4,2,6,4,8]
print(remove_dup(ar))
print(remove_dup(ar1)) 
```

```python
# Q.3. Write a Python class named Circle constructed by a radius and two methods which will compute the area and
# the perimeter of a circle. Add another method which compares the area and perimeter and return the greater of them.
class Circle:
    def __init__(self,n):
        self.n=n
        self.area
    def area(self):
        self.area=3.14*self.n*self.n
        perimeter=2*3.14*self.n
        return f" Area of Circle: {self.area} Perimeter of Circle: {perimeter}"
    def greater(self):
        perimeter=2*3.14*self.n
        if self.area >perimeter :
            return f"area is greater than perimeter"
        else:
            return f"perimeter is greater than area"
a=Circle(187)
print(a.area())
print(a.greater())
    
```

```python
# Q.4. Write a program to generate a 2D in range of -3pi and 3pi  of trigionometric and other mathematical
# function like log, exponent etc with in a single figure of matplot. Use all possible parameters to customize the plot
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')

plt.figure(figsize=(12,10))
plt.subplot(2,2,1)
plt.title('Sin')
plt.xlabel('range -3 to 3')
plt.ylabel('sin of x values')
x=np.arange(-3,4)
y=np.sin(x)
plt.plot(x,y,marker='o')


plt.subplot(2,2,2)

plt.title('Cos')
plt.xlabel('range -3 to 3')
plt.ylabel('cos of x values')
x=np.arange(-3,4)
z=np.cos(x)
plt.plot(x,z,color='cyan')

plt.subplot(2,2,3)

plt.title('log')
plt.xlabel('range -3 to 3')
plt.ylabel('log of x values')
x=np.arange(-3,4)
c=np.log(x)
plt.plot(x,c,color='orange', marker='*')

plt.subplot(2,2,4)

plt.title('Exponential')
plt.xlabel('range -3 to 3')
plt.ylabel('exponential of x values')
x=np.arange(-3,4)
e=np.exp(x)
plt.plot(x,e,color='green', marker='o')
plt.show()
```

```python
# Q.5. Given an integer ( user defined) ,  perform the following conditional actions:

# If  integer is odd, print Weird
# If   integer is even and in the inclusive range of 2 to 5 , print Not Weird
# If  integer is even and in the inclusive range of  6 to 20 , print Weird
# If   integer is even and greater than 20 , print Not Weird

# Sample Input=3
# Sample output=Weired
n=int(input("enter number:"))
if n%2==0:
    if n>=2 and n<=5:
        print("Not Weird")
    if n>=6 and n<=20:
        print('Weird')
    if n>20:
        print('Not Weird')
    
else:
    print('Weird')
```
